```julia
using MLDatasets, Flux
train_data = MLDatasets.MNIST(split=:train)
test_data  = MLDatasets.MNIST(split=:test)

function loader(data; batchsize::Int=1)
    x1dim = reshape(data.features, 28 * 28, :) # reshape 28×28 pixels into a vector of pixels
    yhot  = Flux.onehotbatch(data.targets, 0:9) # make a 10×60000 OneHotMatrix
    Flux.DataLoader((x1dim, yhot); batchsize, shuffle=false)
end
```

```julia
include("graph_ops.jl")

net = declare_net(cross_entropy_loss,
RNN( 14*14 => 64, tanh),
# Dense( 14*14 => 64, tanh),
Dense( 64 => 10, nothing)
)

n_sequences=4

unfold_net!(net,n_sequences)

graph = net.graph
# pprint(graph)
```

```julia
function loss_and_accuracy(net::Network, data)
    (x, y) = only(loader(data; batchsize=length(data)))
    reset!(net.graph)
    feed_with_sequence!(net,
        view(x, 1:196),
        view(x, 197:392),
        view(x, 393:588),
        view(x, 589:784))
    feed_desired_output!(net, y)
    forward!(net.graph)
    ŷ = net.output.output
    acc = round(100 * mean(Flux.onecold(ŷ) .== Flux.onecold(y)); digits=2)
    # Flux.reset!(model)
    # ŷ = model(x[  1:196,:])
    # ŷ = model(x[197:392,:])
    # ŷ = model(x[393:588,:])
    # ŷ = model(x[589:end,:])
    # loss = Flux.logitcrossentropy(ŷ, y)  # did not include softmax in the model
    # acc = round(100 * mean(Flux.onecold(ŷ) .== Flux.onecold(y)); digits=2)
    # (; loss, acc, split=data.split)  # return a NamedTuple
end
```

```julia
using Statistics: mean  # standard library

train_log = []
settings = (;
    eta = 15e-3,
    epochs = 5,
    batchsize = 100,
    lr = 0.05
)
```

```julia


for (x, y) in loader(train_data, batchsize=settings.batchsize)
    feed_with_sequence!(net,
        # view(x, 1:196),
        # view(x, 197:392),
        # view(x, 393:588),
        # view(x, 589:784)
        x[1:196, :],
        x[197:392, :],
        x[393:588, :],
        x[589:784, :]
    )
    feed_desired_output!(net, y)
    init_nodes!(net.graph)
    break
end
```

```julia
using ProgressMeter

loss_in_epoch = 0.0
losses = []



println("before")
@show net.variables[1].output

@time for epoch in 1:settings.epochs
    loss_in_epoch = 0.0
    println("Epoch: ", epoch)
    reset!(graph)
    i = 0
    @time for (x, y) in loader(train_data, batchsize=settings.batchsize)
        i += 1
        feed_with_sequence!(net,
            x[1:196, :],
            x[197:392, :],
            x[393:588, :],
            x[589:784, :]
        )
        feed_desired_output!(net, y)
        currentloss = forward!(graph)
        backward!(graph)
        loss_in_epoch += first(currentloss)
        end
    foreach(x -> x.gradient ./= (length(train_data)/settings.batchsize), net.variables)
adjust!(net.variables, settings.lr)
    @show net.variables[1].gradient
    println("Current loss: ", loss_in_epoch)
    push!(losses, first(loss_in_epoch))
    println("Accuracy: ", loss_and_accuracy(net, train_data))
end
println("after")
@show net.variables[1].output
```
